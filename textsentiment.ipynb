```python
import pandas as pd
import numpy as np
```

Six categories: sadness (0), joy (1), love (2), anger (3), fear (4), and surprise (5).

```python
df=pd.read_csv('/content/textsentiment.csv')
```

```python
df.head()
```

```python
df=df[['text','label']]
df=df.head(2100)
```

Data Cleaning and Preprocessing

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
corpus=[]
for i in range(0,len(df)):
  review = re.sub('^a-zA-Z0-9',' ',df['text'][i])
  review = review.lower()
  review = review.split()

  review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

Creating BoW model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
X.shape
```

```python
X[0].shape
```

```python
y=df['label']
y
```

```python
y.value_counts()
```

Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

For W2V using simple_preprocess from gensim.utils

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt')
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

training W2V from scratch

```python
import gensim
model = gensim.models.Word2Vec(words, window=250, min_count=2)
```

All Vocabulary

```python
model.wv.index_to_key
```

Avg Word2Vec

```python
import numpy as np
```

```python
def avg_w2v(doc):
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
#apply for the entire sentences
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_w2v(words[i]))
```
